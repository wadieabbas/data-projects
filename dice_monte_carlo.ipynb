```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore

num_rolls = 1000000

def dice_roll(n: int):
    di_rolls = np.random.randint(1, 7, n*2)
    roll_pairs = di_rolls.reshape(n, 2)
    
    dice_sum = roll_pairs.sum(axis=1)
    return dice_sum

data_set = dice_roll(num_rolls)
data_set
```

```python
values, counts = np.unique(data_set, return_counts=True)
probabilities = counts / num_rolls * 100

values, counts, probabilities
```

```python
plt.figure(figsize=(10, 6))
plt.bar(values, probabilities)
plt.xlabel('Sum of Dice')
plt.ylabel('Probability (%)')
plt.title('Pair of Dice Rolls Monte Carlo Simulation')
plt.xticks(values)
plt.grid(axis='y')
plt.show()
```
